```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_csv('/NYPD_Arrest_Data__Year_to_Date__20231227.csv')
```

```python
# Group by 'ARREST_BORO' and 'LAW_CAT_CD' and count occurrences
grouped_data = df.groupby(['ARREST_BORO', 'LAW_CAT_CD']).size().reset_index(name='count')

# Pivot the table to have 'ARREST_BORO' as index, 'LAW_CAT_CD' as columns, and 'count' as values
pivot_table = grouped_data.pivot(index='ARREST_BORO', columns='LAW_CAT_CD', values='count').fillna(0)
print(pivot_table)
```

```python
# Filter the data for Brooklyn Borough
brooklyn_data = df[df['ARREST_BORO'] == 'K']

# Group by 'LAW_CAT_CD' and count occurrences for the filtered Brooklyn Data
grouped_data = brooklyn_data.groupby('LAW_CAT_CD').size().reset_index(name='Count')
print(grouped_data)
```

```python
# Create a bar chart
plt.bar(grouped_data['LAW_CAT_CD'], grouped_data['Count'])
plt.xlabel('Law Category')
plt.ylabel('Count')
plt.title('Arrests in Brooklyn Borough by Law Category')
plt.show()
```
